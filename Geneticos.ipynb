```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Especie:


    def __init__(self, pesos : np.array, utilidad_array : np.array,  rep : int, peso_max : int):

        size = len(pesos)
        while True:
            # si se toma la especie o no
            repeticiones = np.random.randint(low=0, high=(rep+1), size=size)

            peso_total = sum(repeticiones * pesos)

            if peso_total <= peso_max:
                break

        self.repeticiones = repeticiones
        self.pesos = pesos
        self.utilidad = utilidad_array
        self.peso_total = peso_total
        self.peso_max = peso_max
        self.rep = rep


    def evaluar_especie(self):

        return sum(self.repeticiones * self.utilidad)

    def cruzar_especie(self, especie_target):


        while True:
            
            operador = np.random.choice([np.maximum, np.bitwise_xor, np.bitwise_and, np.bitwise_or,np.multiply ])
            repeticiones_nueva_especie = operador(self.repeticiones, especie_target.repeticiones)

            peso_total = sum(repeticiones_nueva_especie * self.pesos)

            if peso_total <= self.peso_max:
                break

        especie_mutada = Especie(self.pesos, self.utilidad, self.rep, self.peso_max )
        especie_mutada.repeticiones = repeticiones_nueva_especie

        return especie_mutada
    

    def mutacion(self):
        
        size = len(pesos)
        while True:
            # si se toma la especie o no
            repeticiones = np.random.permutation(self.repeticiones)
            

            peso_total = sum(repeticiones * self.pesos)

            if peso_total <= self.peso_max:
                break
        
        especie_mutada = Especie(self.pesos, self.utilidad, self.rep, self.peso_max )
        especie_mutada.repeticiones = repeticiones

        return especie_mutada

    
```

```python
def nueva_generacion_especies(numero_especies,pesos : np.array, 
                              utilidad_array : np.array,  rep : int, 
                              peso_max : int):
    
    return [ Especie(pesos, utilidad_array, rep, peso_max) for _ in range(numero_especies)]

    
```

```python
especies = nueva_generacion_especies(10,[4, 3, 5, 2], [10, 40, 30, 20],  3, 16)
```
