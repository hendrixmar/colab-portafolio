```
def intersection(image, color, row = 0):

    
    counter = 0
    for r in range(row, a.shape[0]):
        for c in range(a.shape[1]):

            if color == 'black' and image[r][c] in range(0,119):
                return r
            else:
                counter += image[r][c]
        
        if color == 'white' and counter / a.shape[1] < 119:
            return r

            

```

```

def model(image):

    rows = len(image)
    cols = len(image[0])
    output = []

    for r in range(rows):
        output.append(image[:,r])
        output.append(image[r,:])


    return np.array(output).flatten()
```

```

class modelo:
    
    def __init__(self, directory):
        
        self.binarized_training = []
        self.binarized_test = []
        self.mean_vector = []

        for i,img_file in enumerate(os.listdir(directory)):

            img = imageio.imread(f"{directory}/{img_file}",as_gray=True)
            img = model(img)
            
            if i < 45:
                self.binarized_training.append(img < 120)
            else:
                self.binarized_test.append(img < 120)

        self.binarized_test =  np.array(self.binarized_test)
        self.binarized_training = np.array(self.binarized_training)

        for i in range(20000):
            temp = np.mean(self.binarized_training[:,i]) < 0.5
            self.mean_vector.append(temp)
        
        self.mean_vector = np.array(self.mean_vector)

        
    def evaluate(self, image):
        return np.linalg.norm(self.mean_vector^image)

```
