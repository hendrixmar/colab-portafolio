```
from sklearn import datasets 
import numpy as np
import matplotlib.pyplot as plt
from sklearn . preprocessing import StandardScaler
from mlxtend.classifier import Perceptron
from mlxtend.plotting import plot_decision_regions

iris = datasets.load_iris()
X= iris.data
Y= iris.target
```

```
Y
```

```
plt.scatter(X[:50,0], X[:50,1])
plt.scatter(X[50:100,0], X[50:100,1])
plt.scatter(X[100:150,0], X[100:150,1])
```

```
columnas = [0, 3]
XX=X[0:100, columnas] 
YY= iris.target[0:100]
```

```
sc = StandardScaler ()
print( XX.mean(axis=0) ) 
XX = sc.fit_transform(XX) 
print( XX.mean(axis=0) )
```

```
per1 = Perceptron( eta=0.05 , epochs=200 ,
random_seed=1,print_progress=3 )
per1.fit(XX, YY)
plot_decision_regions(XX,YY, clf=per1) 
plt.title("Perceptron")
plt.show()
print(per1.score(XX, YY))
```

```

#1.-Escriba un código que permita mostrar una gráfica similar a la que se muestra 
#en la figura 3.10 con el conjunto de datos Iris.

feats = ["Longitud del Sépalo",  "Ancho del Sépalo", "Longitud del Pétalo", "Ancho del Pétalo"]
plt.rcParams["figure.figsize"] = (18,18)
plt.subplots_adjust(hspace = 0.5, wspace = 0.5)

cont = 1
for i in np.arange(4):
  for j in np.arange(4):
    
    if i!= j:
      plt.subplot(4,4, cont)
      plt.scatter(X[:50, i], X[:50, j])
      plt.scatter(X[50:100, i], X[50:100, j])
      plt.scatter(X[100:150, i], X[100:150, j])
      plt.grid()
    elif i==j:
      plt.subplot(4,4, cont)
      plt.text(0.1, 0.6, feats[i], fontsize=16)
    cont +=1
```
