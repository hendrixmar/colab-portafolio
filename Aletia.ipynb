# ¿Cuál es la probabilidad de salir positivo a la prueba de antígenos, dado que el paciente estuvo en contacto con algún otro caso diagnosticado con SARS CoV-2?
La obtencion de este resultado era mas complejo. Utilice la probabilidad condicional.

```python
muestra_total = df['RESULTADO_ANTIGENO'].isin([2, 1]).sum()
```

```python
resultado_positivo_y_expuesto = ((df['RESULTADO_ANTIGENO'] == 1) & (df['OTRO_CASO'] == 1)).sum() / muestra_total
```

```python
(resultado_positivo_y_expuesto / ((df['OTRO_CASO'] == 1).sum() / muestra_total)) *100
```
