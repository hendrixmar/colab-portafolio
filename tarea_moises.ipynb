#4. Repita el ejercicio anterior pero aplicando el algoritmo de Rosenblatt, es decir, no use la biblioteca MLxtend.

```
np.random.seed(0)
W = np.random.uniform(low=-0.2, high=0.2, size=3)
signo = lambda x : 1.0 if x >= 0 else -1.0

n = 0.25



class Perceptron(object):

    def __init__(self, no_of_inputs=2, threshold=200, learning_rate=0.25):
        self.threshold = threshold
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_inputs + 1)
           
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation > 0:
          activation = 1
        else:
          activation = 0            
        return activation

    def train(self, training_inputs, labels):
        for _ in range(self.threshold):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs
                self.weights[0] += self.learning_rate * (label - prediction)


perce = Perceptron()

perce.train(XX,YY)
perce.weights
output = []
for i in XX:
    output.append(perce.predict(i))

output = np.array(output)
print(sum(YY == output)/len(XX))
```
