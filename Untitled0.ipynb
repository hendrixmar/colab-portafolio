```
import random
import math
import matplotlib.pyplot as plt


# random angle
def generate_circle_points(number_points):
    
    # radius of the circle
    circle_r = 100
    # center of the circle (x, y)
    circle_x = 200
    circle_y = 200

    points = []
    r = circle_r * math.sqrt(random.random())
    # calculating coordinates
    for _ in range(number_points):
        alpha = 2 * math.pi * random.random()
        # random radius
        
        x = r * math.cos(alpha) + circle_x
        y = r * math.sin(alpha) + circle_y
        
            
        points.append((x,y))
    
    return points


def generate_square_points(number_points):
    
    vertices = [(0,0),(0,50),(50,50),(50,0)]
    points = []
    for _ in range(number_points):
        
        vertex_a = random.randint(0,3)
        vertex_b = (vertex_a + 1) % 4
            
        
        if vertices[vertex_a][0] == vertices[vertex_b][0]:
            
            new_point = (vertices[vertex_a][0], random.random()*50)
        else:
            new_point = (random.random()*50, vertices[vertex_a][1])
        points.append(new_point)
    
    print(points)
    return points


generate_square_points(50)

"""
plt.scatter(*zip(*generate_square_points(50)))
plt.ylabel('Distance')
plt.xlabel('Generation')
plt.show()
   """      

```
