```python
import csv
import re
```

```python
def load_csv(path: str = "world_population.csv", option: str = 'r'):
    with open(path, option) as file:
        for row in csv.reader(file):
            yield row

        
```

```python
def convert_csv_to_json(file_stream,
                        column_names=[]):
    return (dict(zip(column_names, row))
            for row in file_stream)

# function 2
def transform_dict(dict_row, column_name='Population'):

    return {
        **{key: value for key, value in dict_row.items() if not re.match(r'\d{4} Population', key)},
        **{column_name: [value for key, value in dict_row.items() if re.match(r'\d{4} Population', key)]}
    }

# function 3
def segregate_country_population(dict_row, filter_function):
    return {key: value for key, value in dict_row.items() if filter_function(key, value)}

#function 4
def country_biggest_population(dict_rows):
    return ((row.get('CCA3'), row.get('World Population Percentage')) 
            for row in dict_rows if float(row.get('World Population Percentage')) >= 1.0)

avg = lambda seq: sum(int(_) for _ in seq) / len(seq)
#function 5
def country_avg_population(dict_row, key= 'Population'):
    return {**dict_row, **{key:avg(dict_row.get(key))}}

# function 6
def country_continent_map(dict_rows):
    return ((row.get('Country/Territory'),row.get('Continent')) for row in dict_rows)

# function 7
def continents_map(dict_rows):
    return (continent for _, continent in country_continent_map(dict_rows))

# function 8 - 13
def filter_country_population(dict_rows, columns_included, continent):
    return ( tuple(json.get(key) for key in columns_included) 
                for json in dict_rows if json.get('Continent') in continent)
    
# function 16 - 17
def filter_country_continent(dict_rows, filter):
    return ( row for row in dict_rows if filter(row))


```

```python
file_stream = load_csv()

columns_names = next(file_stream)
print(len(columns_names))
result = list(convert_csv_to_json(file_stream, columns_names))
result = (transform_dict(row) for row in result)

filter_country_continent(result, lambda x: x.get('World Population') >= 0.2 )
filter_country_continent(result, lambda x: x.get('World Population') >= 0.2 )
```
