3.1- Modelado:
- Con cada una de las imágenes se construirá un vector al concatenar una columna de la
imagen y el vector transpuesto de cada renglón; primero un vector de la columna y
enseguida el vector correspondiente al primer renglón, posteriormente la segunda columna
seguida del segundo renglón, y así hasta terminar con toda la imagen.
- El modelo final se obtendrá de la media entre las primeras 15 letras (base de datos de
aprendizaje) al igual que en el ejercicio anterior.

```

class modelo:
    
    def __init__(self, directory):
        
        self.binarized_training = []
        self.binarized_test = []
        self.mean_vector = []

        for i,img_file in enumerate(os.listdir(directory)):

            img = imageio.imread(f"{directory}/{img_file}",as_gray=True)
            img = model(img)
            
            if i < 45:
                self.binarized_training.append(img < 120)
            else:
                self.binarized_test.append(img < 120)

        self.binarized_test =  np.array(self.binarized_test)
        self.binarized_training = np.array(self.binarized_training)

        for i in range(20000):
            temp = np.mean(self.binarized_training[:,i]) < 0.5
            self.mean_vector.append(temp)
        
        self.mean_vector = np.array(self.mean_vector)

        
    def evaluate(self, image):
        return np.linalg.norm(self.mean_vector^image)

```

```


modelo_1 = modelo("images") # Manoescrita
modelo_2 = modelo("images_2") # imprenta
print(modelo_1.mean_vector)
print(modelo_2.mean_vector)
```

```
print("Modelo 1")
print("-----manoescritas")
for image in modelo_1.binarized_test:
    print(modelo_1.evaluate(image))

print("-----imprenta")

for image in modelo_2.binarized_test:
    print(modelo_1.evaluate(image))

print("Modelo 2")
print("------manoescrita")
for image in modelo_1.binarized_test:
    print(modelo_2.evaluate(image))
    
print("-----imprenta")
for image in modelo_2.binarized_test:
    print(modelo_2.evaluate(image))



```
